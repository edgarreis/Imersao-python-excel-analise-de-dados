Criando Gráficos de Candlestick com Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
dados = yf.download('PETR4.SA', start = '2023-01-01', end='2023-12-31')
dados
```

```python
dados.columns
```

```python
dados.columns = ['Abertura', 'Máximo', 'Mínimo', 'Fechamento', 'Fech_Ajust', 'Volume']
```

```python
dados = dados.rename_axis('Data')
dados
```

```python
dados['Fechamento'].plot(figsize=(10,6))
plt.title('Variação do Preço por Data', fontsize=16)
plt.legend(['Fechamento'])
```

```python
df = dados.head(60).copy()
# Convertendo o index em uma coluna de data
df['Data'] = df.index
# TRANSFORMANDO DATAS EM NUMEROS
# Convertendo as datas para o formato numérico de matplotlib
# Isso é necessário para que o Matplotlib possa plotar as datas correstamente no gráfico
df['Data'] = df['Data'].apply(mdates.date2num)
df
```

```python
fig, ax = plt.subplots(figsize=(15, 8))

# Vamos definir a largura dos candles no gráfico
width = 0.7

for i in range(len(df)):
    # Determinando a cor do candle (verde=subiu, vermelho=desceu)
    if df['Fechamento'].iloc[i] > df['Abertura'].iloc[i]:
      color = 'green'
    else:
      color = 'red'

    # Desenhando as linhas verticais do candle (Máximo e Mínimo)
    # Usando o 'ax.plot' para desernhar uma linha vertical
    ax.plot([df['Data'].iloc[i], df['Data'].iloc[i]],
            [df['Mínimo'].iloc[i], df['Máximo'].iloc[i]],
            color=color,
            linewidth=1)

    # Desenha os retangulos de Abertura e Fechamento do dia
    ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width/2, min(df['Abertura'].iloc[i], df['Fechamento'].iloc[i])),
                               width,
                               abs(df['Fechamento'].iloc[i] - df['Abertura'].iloc[i]),
                               facecolor=color))

# Médias Móveis
df['MA7'] = df['Fechamento'].rolling(window=7).mean()
df['MA14'] = df['Fechamento'].rolling(window=14).mean()

# Plotando as médias móveis
ax.plot(df['Data'], df['MA7'], color='orange', label='Média Móvel 7 dias')
ax.plot(df['Data'], df['MA14'], color='yellow', label='Média Móvel 14 dias')
ax.legend()

# Formatando eixo x
ax.xaxis_date()   # Config. para o Matplotlib
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Adicionando título e rótulo para os eixos x e y
plt.title('Grafico de Candlestick - PETR4.SA com matplotlib')
plt.xlabel('Data')
plt.ylabel('Preço')

# Adicionando uma grade
plt.grid(True)

# Exibindo o gráfico
plt.show()
```
