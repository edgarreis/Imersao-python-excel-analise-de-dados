```python
!pip install gspread

import pandas as pd
import plotly.express as px
import gspread
import locale
from google.colab import auth
```

Importando os dados do Excel

```python
df_principal = pd.read_excel("/content/Tabela_de_Acoes.xlsx", sheet_name="Principal") #Dataflame
df_principal.head(4)
```

Transformando as Planilhas em DataFrame

```python
df_Total_de_acoes = pd.read_excel("/content/Tabela_de_Acoes.xlsx", sheet_name="Total_de_acoes") #Dataflame
df_Total_de_acoes.head(3)
```

Transformando as Planilhas em DataFrame

```python
df_Ticker = pd.read_excel("/content/Tabela_de_Acoes.xlsx", sheet_name="Ticker") #Dataflame
df_Ticker.head(3)
```

Transformando as Planilhas em DataFrame

```python
df_ChatGPT = pd.read_excel("/content/Tabela_de_Acoes.xlsx", sheet_name="ChatGPT") #Dataflame
df_ChatGPT.head(3)
```

Deletando as Colunas Não Necessárias

```python
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()
df_principal
```

Renomeando as Colulas

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

Calculos da Variação Percentual e Valor Inicial de cada Ação

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal
```

Incorporando a Quantidade Téorica de outra tabela para dentro do DataFrame_Principal, similar ao **PROCV** do Excel

```python
df_principal = df_principal.merge(df_Total_de_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

Deletando a Coluna Código

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

Calculando a Variação em Reais

```python
df_principal['variacao_rs'] = ( df_principal['valor_final'] - df_principal['valor_inicial'] ) * df_principal['Qtde. Teórica']
df_principal
```

Formatando os Numeros para vizualização

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

Converte para INT

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={"Qtde. Teórica":"qtd_teorica"}).copy()
df_principal
```

```python
df_principal['resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_ChatGPT, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={"Idade (anos)":"idade"}).copy()
df_principal
```

```python
df_principal['cat_idade'] = df_principal['idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos que 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
# Calcular os valores desejados
maior = df_principal['variacao_rs'].max()
menor = df_principal['variacao_rs'].min()
media = df_principal['variacao_rs'].mean()
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao_rs'].mean()    # Média Se
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao_rs'].mean()  # Média Se

# Exibir os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

**Graficos Para Análise**

```python
fig = px.bar(df_analise_saldo, x='variacao_rs', y='resultado', text='variacao_rs', title='Variação Reais por Resultado')
# Atualizar a formatação do texto das barras para exibir os valores com ponto como separador de milhares e duas casas decimais
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

fig.show()
```

```python
# Criar o gráfico de pizza
fig = px.pie(df_analise_segmento, values='variacao_rs', names='Segmento', title='Total de Vendas por Segmento')

# Exibir o gráfico
fig.show()
```

```python
# Ordenar o DataFrame pelo número de pessoas em ordem decrescente
df_analise_segmento_sorted = df_analise_segmento.sort_values(by='variacao_rs', ascending=False)

# Criar o gráfico de barras
fig = px.bar(df_analise_segmento_sorted, x='Segmento', y='variacao_rs', title='Quantidade por Faixa de Idade')

# Exibir o gráfico
fig.show()
```

```python
df_analise_cat_idade = df_principal.groupby('cat_idade')['variacao_rs'].sum().reset_index()
df_analise_cat_idade

fig = px.bar(df_analise_cat_idade, x='cat_idade', y='variacao_rs', text='variacao_rs', title='Variação Reais por Categoria de Idade')
fig.show()
```
